```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos para la varilla de hierro
I = np.array([0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])
minimuns_iron = np.array([30, 37, 41, 45, 52, 59, 64, 67, 69, 75, 84])

# Definir una función de ajuste (en este caso, un polinomio de grado 2)
def func(x, a, b, c):
    return a * x**2 + b * x + c

# Realizar el ajuste de curva para los datos de hierro
params_iron, covariance_iron = curve_fit(func, I, minimuns_iron)

# Parámetros ajustados para el hierro
a_iron, b_iron, c_iron = params_iron

# Crear un rango de valores de I para la curva ajustada
I_fit = np.linspace(0, 5, 100)

# Calcular los valores ajustados correspondientes para el hierro
minimuns_fit_iron = func(I_fit, a_iron, b_iron, c_iron)

# Calcular los residuos para los datos de hierro
residuos_iron = minimuns_iron - func(I, a_iron, b_iron, c_iron)

# Crear una figura con dos subplots en una columna con dos filas
plt.figure(figsize=(8, 12))

# Primer subplot: Gráfico de datos y ajuste para el hierro
plt.subplot(2, 1, 1)
plt.title("I vs Número de mínimos Fe")
plt.xlabel("I (A)")
plt.ylabel("Número de mínimos observados")
plt.scatter(I, minimuns_iron, color="black", label="#mínimos")
plt.plot(I_fit, minimuns_fit_iron, color="red", label="Ajuste polinómico")
plt.grid(True)
plt.legend()

# Segundo subplot: Gráfico de residuos para el hierro
plt.subplot(2, 1, 2)
plt.title("Residuales")
plt.xlabel("Índice de Datos")
plt.ylabel("Residuos")
plt.scatter(range(len(I)), residuos_iron, color="black")
plt.axhline(y=0, color="red", linestyle="--", label="Línea base")
plt.grid(True)
plt.legend()

# Ajustar la disposición de los subplots para evitar superposiciones
plt.tight_layout()

# Mostrar la figura combinada
plt.show()

# Mostrar los parámetros ajustados para el hierro
print("Parámetros del ajuste para el hierro:")
print(f"a_iron = {a_iron}")
print(f"b_iron = {b_iron}")
print(f"c_iron = {c_iron}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
I = np.array([0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])
minimuns = np.array([10, 16, 24, 30, 35, 39, 44, 47, 51, 55, 59])

# Definir una función de ajuste (en este caso, un polinomio de grado 2)
def func(x, a, b, c):
    return a * x**2 + b * x + c

# Realizar el ajuste de curva
params, covariance = curve_fit(func, I, minimuns)

# Parámetros ajustados
a, b, c = params

# Crear un rango de valores de I para la curva ajustada
I_fit = np.linspace(0, 5, 100)

# Calcular los valores ajustados correspondientes
minimuns_fit = func(I_fit, a, b, c)

# Calcular los residuos
residuos = minimuns - func(I, a, b, c)

# Crear una figura con dos subplots en una columna con dos filas
plt.figure(figsize=(8, 12))

# Primer subplot: Gráfico de datos y ajuste
plt.subplot(2, 1, 1)
plt.title("I vs Número de mínimos Ni")
plt.xlabel("I (A)")
plt.ylabel("Número de mínimos observados")
plt.scatter(I, minimuns, color="black", label="# mínimos")
plt.plot(I_fit, minimuns_fit, color="red", label="Ajuste polinómico con n=2")
plt.grid(True)
plt.legend()

# Segundo subplot: Gráfico de residuos
plt.subplot(2, 1, 2)
plt.title("Residuales")
plt.xlabel("Índice de Datos")
plt.ylabel("Residuos")
plt.scatter(range(len(I)), residuos, color="black")
plt.axhline(y=0, color="red", linestyle="--", label="Línea base")
plt.grid(True)
plt.legend()

# Ajustar la disposición de los subplots para evitar superposiciones
plt.tight_layout()

# Mostrar la figura combinada
plt.show()

# Mostrar los parámetros ajustados
print("Parámetros del ajuste:")
print(f"a = {a}")
print(f"b = {b}")
print(f"c = {c}")
```
